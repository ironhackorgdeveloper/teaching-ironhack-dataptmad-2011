# Web Scraping

## 1. Introduction

Web scraping refers to the automatic extraction of information from a web page. This information is often a page's content, but it can also include information in the page's headers, links present on the page, or any other information embedded in the page's HTML. Because of this, scraping has become one of the most popular ways to extract data from the web. 

_"With basic knowledge of HTML and the help of a few Python libraries, you can obtain information from just about any page on the Internet"_

## 3. Scraping a simple website

```python
import requests

url = 'https://datatau.net'
html = requests.get(url).content
html[:5000]
```

### 3.1 Beautiful Soup

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'html.parser')

soup
```

```python
table = soup.find('table', {'class': 'itemlist'})

table
```

```python
table.find_all('a', {'class': 'storylink'})
```

```python
first_link_html = table.find_all('a', {'class': 'storylink'})[0]

first_link_html
```

```python
first_link_html.text
```

```python
first_link_html.attrs
```

```python
first_link_html.attrs['href']
```

```python
titles = [x.text for x in table.find_all('a', {'class': 'storylink'})]
urls = [x.attrs['href'] for x in table.find_all('a', {'class': 'storylink'})]
publications = [x.text for x in table.find_all('span', {'class': 'sitestr'})]
scores = [x.text for x in table.find_all('span', {'class': 'score'})]
authors = [x.text for x in table.find_all('a', {'class': 'hnuser'})]
time_agos = [x.text for x in table.find_all('a', {'class': 'age'})]

info_dict = {
    'title': titles,
    'url': urls,
    'publication': publications,
    'score': scores,
    'author': authors,
    'time_ago': time_agos,
}

info_dict
```

```python
import pandas as pd

df = pd.DataFrame(info_dict)

df.head()
```

```python
import re

test_str = '\n 2 points'

parsed_int = int(re.sub(r'\D', '', test_str))

parsed_int
```

```python
import re

# df['score'] = df['score'].str.replace(r'\D', '').astype(int)
df['score'] = df['score'].apply(lambda x: int(re.sub(r'\D', '', x)))

df.head()
```
