# data aggregation (with pandas)

## 1. introduction

* groupby is one of the most interesting methods of pandas df
* it is not easy to understand groupby objects at first sight, so repeated practice is very important here

## 2. data loading

Famous dataset of classic vehicles...

```python
import pandas as pd
import numpy as np
```

Load `.csv` inside zip file...

```python
from zipfile import ZipFile

z = ZipFile('./data/vehicles.zip')

z.infolist()
```

```python
vehicles = pd.read_csv(z.open('vehicles/vehicles.csv'))
```

```python
vehicles.head().T
```

```python
vehicles.info(memory_usage='deep')
```

```python
vehicles.describe(include='all')
```

```python
vehicles.describe(include='all')
```

## 3. groupby object

```python
vehicles.groupby('Transmission')
```

nothing to do with this object itself, let's explore it deeper...

## 4. aggregations

### single aggregation function, aggregation by a single column

```python
vehicles.groupby('Transmission') \
['Highway MPG', 'City MPG', 'Combined MPG'].mean()
```

### aggregation by multiple columns, single aggregation function

```python
vehicles.groupby(['Fuel Type', 'Cylinders']) \
    ['CO2 Emission Grams/Mile'].mean().sort_values(ascending=True)
```

### aggregation using agg method

```python
vehicles.groupby('Fuel Type')['Combined MPG', 'Highway MPG'].agg(['min', 'max', 'mean', 'median', 'std'])
```

### specific aggregation for columns

```python
vehicles.groupby('Fuel Type').agg(min_mpg_combined=('Combined MPG', 'min'),
                                  max_mpg_highway=('Highway MPG', 'max'))
```

### custom aggregations

```python
from scipy import stats
```

```python
stats.mode([1, 2, 3, 3, 4])
```

```python
def agg_custom_function(col):
    return np.max(col) / np.min(col)
    # return stats.mode(col)[0]
```

```python
vehicles.groupby('Transmission')["Combined MPG"].agg(agg_custom_function)
```
