```python
import pandas as pd
```

## 1. CSV

```python
# import comma-separated variable file
df = pd.read_csv('data/vehicles.csv')

df.head()
```

```python
pd.read_csv('data/vehicles_pipe.txt')
```

```python
# import tab-delimited file
df_tab = pd.read_csv('data/vehicles_tab.txt', sep='\t')

df_tab.head()
```

```python
# import pipe-delimited file
# sep and delimiter are equivalents
df_pipe = pd.read_csv('data/vehicles_pipe.txt', delimiter='|')

df_pipe.head()
```

## 5. Database

```python
!conda install -y sqlalchemy
```

```python
from sqlalchemy import create_engine
```

```python
sqlitedb_path = '../../week_03/sql_introduction/data/chinook.db'
```

```python
engine = create_engine(f'sqlite:///{sqlitedb_path}')
```

```python
# list all tables in database
pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", engine)
```

```python
# query one of them
df_sqlite = pd.read_sql_query("select * from customers", engine)

df_sqlite.head()
```

```python
two_lines_query = """
select * from customers
where FirstName='Leonie'
"""
```

```python
df_sqlite = pd.read_sql_query(two_lines_query, engine)

df_sqlite.head()
```

```python
df_sqlite.to_sql('test_write', engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", engine)
```

```python
drop_query = 'drop table test_write'

# pd.read_sql_query(drop_query, engine)

with engine.begin() as conn:
    conn.execute(drop_query)
```
